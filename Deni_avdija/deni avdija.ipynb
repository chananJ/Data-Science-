- BeautifulSoup is used to parse HTML content, making it easier to extract specific elements and attributes from HTML documents.


- Here, BeautifulSoup helps us locate the table containing the game log data. 
- We search for an HTML element with the ID "pgl_basic" using `soup.find(id="pgl_basic")`. 
- Once found, this element represents the game log table.


```python
table_2021=soup.find(id="pgl_basic")
```

```python
#table_2021A=pd.read_html(str(table_2021))[0]
```

- Based on this concept: `table_2021A = pd.read_html(str(table_2021))[0]` will iterate through all the HTML and retrieve the data.


```python
dfs=[]
for year in years:
    with open("C:\\Users\\user\\deni_avdija\\{}.html".format(year),encoding="utf-8") as f :
        page=f.read()
        soup = BeautifulSoup(page, "html.parser")
        table=soup.find(id="pgl_basic")
        table=pd.read_html(str(table))[0]
        dfs.append(table)
```

```python
i=0
writer = pd.ExcelWriter('output.xlsx')

for df in dfs:
    name="deni"+str(i)
    print(name)
    i=i+1
    df.to_excel(writer, sheet_name=name)
    
writer.save()
```
